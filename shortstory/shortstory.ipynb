# Study of cost of insurance rate
## Somker and non-smoker distribution of data

```python
plt.xlabel('Non-Smoker = 0, Smoker = 1')
plt.ylabel('Frequency')
plt.title('Smoker Distribution')
df.smoker.value_counts().plot(kind='bar')
plt.show()
```

```python
sns.catplot(x='smoker', y='charges', data=df)
plt.show()
```

```python
sns.catplot(x='smoker', y='charges', hue='sex', kind='violin', data=df[df['age'] == 18], legend=True)
plt.show()
```

## Medical cost distribution in various region

```python
sns.boxplot(x=df.region, y=df.charges, data=df)
plt.title("medical charges per region")
plt.show()
display()
```
